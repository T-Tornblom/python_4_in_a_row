```python
def who_starts(name1, name2):
    t = 0

    value, card ,s= random_card()
    value2, card2 ,s2= random_card()
    if value > value2 or (value == value2 and s>s2):
        print(f" {name1} {value} <-- Starts \n{card} {name2} {value2} \n{card2}")
        return_name1 = name1
        return_name2 = name2
    else:
        print(f" {name1} {value} \n{card} {name2} {value2} <-- Starts \n{card2}")
        return_name1 = name2
        return_name2 = name1
    while t == 0:
        t = input("Are you ready to start?")
    return  return_name1, return_name2
```

```python
def who_is_who():
    
    player1_name = input("First player - What is your name?")

    player2_name = input("Second player - What is your name?")
  
    return player1_name , player2_name
```

# Game Logic 

```python
from IPython.display import clear_output
from colorama import Fore
from random import randint
from time import sleep

while True:
    
    board = reset_board()
    name1, name2 = who_is_who()
    refresh_screen()
    player1_name, player2_name = who_starts(name1, name2)
    
    player1_marker="X"
    player2_marker="O" 
    turn = "Player 1"
    game_on = True
    while game_on:
        
        #Player 1 turn
        if turn == "Player 1":
            
            display_board(board) # Display the current state of the board
            print("It's yout turn" , player1_name ," you are " , player1_marker )
            
            position = player_inpt(board, player1_marker) # Player inputs where they want to drop their marker
            
            drp_lgc(player1_marker,position) # Calculates where the marker should end up.
            
            if win_check(board, player1_marker):
                
                player_x_win(board, player1_name)
                
                game_on = False
            else:
                if full_tp_row(board):
                    
                    print_draw()
                    break
                else:
                    turn = "Player 2"
        
        # Player 2 turn
        else:
            
            display_board(board)
            
            print("It's yout turn" , player2_name ," you are " , player2_marker )
            
            position = player_inpt(board, player2_marker)
            
            drp_lgc(player2_marker,position)
            
            if win_check(board, player2_marker):
                
                player_O_win(board, player2_name)
                
                game_on = False
            else:
                if full_tp_row(board):
                    
                    print_draw()
                    break
                else:
                    turn = "Player 1" 
    if not replay():
        break
```
